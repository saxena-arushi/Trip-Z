```python
import pandas as pd
import joblib
df = pd.read_excel('Trip-Z Survey Form (Responses) (1).xlsx')
df = df.drop(['First Name', 'Last Name'], axis=1)
df = df.applymap(lambda x: x.split(','))
df.head()
```

```python
size = df.shape[0]
dataset = []
for i in range(size):
    row_list = df.loc[i, :].values.flatten().tolist()
    flat_list = [item for sublist in row_list for item in sublist]
    dataset.append(flat_list)
    
print(dataset)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)*1
df 
```

```python
# dataset = pd.read_csv('test.csv')
# X = dataset.iloc[:, :].values
```

```python
import numpy as np
from sklearn.cluster import KMeans
X = df.iloc[:, :].values
# # X = np.array(df)
# X = df.values


```

```python
kmeans= KMeans(n_clusters=5)
kmeans.fit(X)
identified_clusters = kmeans.fit_predict(X)
identified_clusters = list(identified_clusters)
cluster_labels = kmeans.predict(X)
df['cluster'] = cluster_labels
df
```
